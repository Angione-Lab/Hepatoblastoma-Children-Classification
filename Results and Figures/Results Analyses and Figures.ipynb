# Visualisation Analyses - Weight analyses

```python
data_path = "path\Feature weights\\"
FVA_weights = pd.read_csv(data_path + "FVA_weights.csv")
FVA_clinical_weights = pd.read_csv(data_path + "FVA_clinical_weights.csv")
FVA_genes_weights = pd.read_csv(data_path + "FVA_genes_weights.csv")
FVA_all_weights = pd.read_csv(data_path + "FVA_all_weights.csv")

pathways_weights = pd.read_csv(data_path + "pathways_weights.csv")
pathways_clinical_weights = pd.read_csv(data_path + "pathways_clinical_weights.csv")
pathways_genes_weights = pd.read_csv(data_path + "pathways_genes_weights.csv")
pathways_all_weights = pd.read_csv(data_path + "pathways_all_weights.csv")

genes_weights = pd.read_csv(data_path + "genes_weights.csv")
genes_clinical_weights = pd.read_csv(data_path + "genes_clinical_weights.csv")
genes_FVA_weights = pd.read_csv(data_path + "genes_FVA_weights.csv")
genes_all_weights = pd.read_csv(data_path + "genes_all_weights.csv")

def histFormat(dataset):
    if len(dataset.columns) == 4: # fluxes
        dataset = dataset[dataset.columns[1:]]
        omic = "FVA"
    else:
        omic = dataset.columns[0]
    dataset["Omic"] = omic
    dataset.columns = ["Name", "Weight", "Omic combination", "Omic"]
    return dataset
```

```python
FVA_weights = histFormat(FVA_weights)
FVA_clinical_weights = histFormat(FVA_clinical_weights)
FVA_genes_weights = histFormat(FVA_genes_weights)
FVA_all_weights = histFormat(FVA_all_weights)

pathways_weights = histFormat(pathways_weights)
pathways_clinical_weights = histFormat(pathways_clinical_weights)
pathways_genes_weights = histFormat(pathways_genes_weights)
pathways_all_weights = histFormat(pathways_all_weights)

genes_weights = histFormat(genes_weights)
genes_clinical_weights = histFormat(genes_clinical_weights)
genes_FVA_weights = histFormat(genes_FVA_weights)
genes_all_weights = histFormat(genes_all_weights)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

sns.histplot(FVA_weights, x="Weight", ax=axes[0, 0])
sns.histplot(FVA_genes_weights, x="Weight", ax=axes[0, 1])
sns.histplot(FVA_clinical_weights, x="Weight", ax=axes[1, 0])
sns.histplot(FVA_all_weights, x="Weight", ax=axes[1, 1])

fig.suptitle("Weight distributions for FVA fluxes")
axes[0, 0].title.set_text("Only fluxes")
axes[0, 1].title.set_text("Fluxes and genes")
axes[1, 0].title.set_text("Fluxes and clinical")
axes[1, 1].title.set_text("Fluxes, genes and clinical")

plt.savefig(data_path + "FVA_weight_distributions.pdf")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

sns.histplot(pathways_weights, x="Weight", ax=axes[0, 0])
sns.histplot(pathways_genes_weights, x="Weight", ax=axes[0, 1])
sns.histplot(pathways_clinical_weights, x="Weight", ax=axes[1, 0])
sns.histplot(pathways_all_weights, x="Weight", ax=axes[1, 1])

fig.suptitle("Weight distributions for FVA pathways")
axes[0, 0].title.set_text("Only fluxes")
axes[0, 1].title.set_text("Fluxes and genes")
axes[1, 0].title.set_text("Fluxes and clinical")
axes[1, 1].title.set_text("Fluxes, genes and clinical")

plt.savefig(data_path + "pathway_weight_distributions.pdf")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

sns.histplot(genes_weights, x="Weight", ax=axes[0, 0])
sns.histplot(genes_FVA_weights, x="Weight", ax=axes[0, 1])
sns.histplot(genes_clinical_weights, x="Weight", ax=axes[1, 0])
sns.histplot(genes_all_weights, x="Weight", ax=axes[1, 1])

fig.suptitle("Weight distributions for genes")
axes[0, 0].title.set_text("Only genes")
axes[0, 1].title.set_text("Genes and fluxes")
axes[1, 0].title.set_text("Genes and clinical")
axes[1, 1].title.set_text("Genes, fluxes and clinical")

plt.savefig(data_path + "genes_weight_distributions.pdf")
```
